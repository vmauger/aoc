```python
# day 2
path = "C:\\Users\\Valentin\\Documents\\aoc\\data_day2.csv"
file = open(path,"r")

#Part 1
mapping = {
    0:3, 
    1:6,
    -2:6,
    -1:0,
    2:0
}
value = 0
for round in file.readlines():
    round = round.rstrip('\n').split(' ')
    diff = ((ord(round[1].lower())-119))-((ord(round[0].lower())-96))
    round.append(mapping[diff]+((ord(round[1].lower())-119)))
    value += round[-1]
print("First Part:",value)

#Part 2
file.seek(0)
value = 0
mapping2 = {
    ('A','X'):3,
    ('A','Y'):4,
    ('A','Z'):8,
    ('B','X'):1,
    ('B','Y'):5,
    ('B','Z'):9,
    ('C','X'):2,
    ('C','Y'):6,
    ('C','Z'):7
}
for round in file.readlines():
    round = round.rstrip('\n').split(' ')
    round.append(mapping2[tuple(round)])
    value += round[-1]
print("Second Part:",value)
file.close()
```

```python
# day 3
def get_priority(letter):
    priority_list = list('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ')
    return priority_list.index(letter)+1


def read_data():
    global data
    path = "C:\\Users\\Valentin\\Documents\\aoc\\data_day3.csv"
    with open(path) as f:
        data = f.read()
    data = data.strip('\n\n')
    data = data.split('\n')

def first_part():
    value = 0
    for ruck in data:
        set_1 = {i for i in list(ruck[:int(len(ruck)/2)])}
        set_2 = {i for i in list(ruck[int(len(ruck)/2):])}
        intersection = list(set_1.intersection(set_2))
        if intersection !=[]:
            value += get_priority(intersection[0])
    print(value)


def second_part():
    value = 0
    sub_list = [data[ruck:ruck+3] for ruck in range(0,len(data),3)]
    for group in sub_list:
        set_1 = {i for i in list(group[0])}
        set_2 = {i for i in list(group[1])}
        set_3 = {i for i in list(group[2])}

        intersection = list(set_1.intersection(set_2.intersection(set_3)))
        if intersection !=[]:
            value += get_priority(intersection[0])
    print(value)

read_data()
first_part()
second_part()
```

```python
#day 5
def read_data():
    global cargo, hauls
    path = "C:\\Users\\Valentin\\Documents\\aoc\\data_day5.csv"
    with open(path) as f:
        data = f.read()
    cargo, hauls = data.split('\n\n')

read_data()
hauls = hauls.strip('\n\n')
hauls = hauls.split('\n')
for i, actions in enumerate(hauls):
    actions = actions.split(' ')
    hauls[i] = actions


cargo = cargo.split('\n')
cargo = cargo[::-1]
stacks = {}
for depth in cargo:
    depth = [depth[i] for i in range(1,len(cargo[0]),4)]    
    for i, crate in enumerate(depth):
        if int(i+1) not in stacks.keys():
            stacks[i+1] = list(crate)
        else:
            stacks[i+1].append(crate)
for key, stack in stacks.items():
    stack = stack[1:]
    if " " in stack:
        stack = stack[:stack.index(' ')]
    stacks[key] = stack


for actions in hauls:
    temp_stack = stacks[int(actions[3])][int(actions[1])*-1:]
    stacks[int(actions[3])] = stacks[int(actions[3])][:int(actions[1])*-1]
    # temp_stack = temp_stack[::-1] #Comment it you want part 2 lol
    stacks[int(actions[5])] = stacks[int(actions[5])] + temp_stack

result = ''
for stack in stacks.values():
    result = result + stack[-1]
print(result)
```

```python
#day 7
class Folder:
    def __init__(self, path):
        self.path = path
        if path[path.rstrip('/').rfind('/')+1:path.rfind('/')] == '':
            self.name = '/'
        else:
            self.name = path[path.rstrip('/').rfind('/')+1:path.rfind('/')]+'/'
        self.parent_path = path[:path.rstrip('/').rfind('/')]+'/'
        self.content = []

    def __str__(self):
        return self.path

    def __eq__(self, other):
        return self.path == other.path

    def __int__(self):
        x = 0
        for items in self.content:
            x += items[1]
        return x

    def __repr__(self):
        return f"[{str(self)}, {int(self)}]"

    def __getitem__(self,items):
        if items == 1:
            return int(self)

    def add_file(self, filename, size, override=False):
        new_file = [filename, int(size)]
        self.content.append(new_file)

    def add_directory(self, folder, override=False):
        self.content.append(folder) 


def read_data():
    global log
    path = "C:\\Users\\Valentin\\Documents\\aoc\\data_day7.csv"
    with open(path) as f:
        log = f.read()
    log = log.strip('\n\n')
    log = log.split('\n')

read_data()

folder = []
folder.append(Folder('/'))
path = ''
for line in log:
    line = line.split(' ')
    if line[0] == '$' and line[1] == 'cd':
        if line[2][0] == '/':
            path = line[2]
        elif line[2] == '..':
            path = path[:path.rstrip('/').rfind('/')]+'/'
        else:
            path += line[2]+'/'
    elif line[0] == 'dir':
        folder.append(Folder(str(path)+line[1]+'/'))
        folder[folder.index(Folder(path))].add_directory(folder[folder.index(Folder(path+line[1]+'/'))])
    elif line[0] == '$' and line[1] == 'ls':
        pass
    else:
        folder[folder.index(Folder(path))].add_file(str(line[1]), int(line[0]))

value = 0
for folders in folder:
    if int(folders)<=100000:
        value += int(folders)
print(value)


total_space = 70000000
used_space = int(folder[0])
needed_space = 30000000
to_free_space = needed_space-(total_space-used_space)
folder_volumes = []
for folders in folder:
    if int(folders) > to_free_space:
        folder_volumes.append(int(folders))
    folder_volumes.sort()
print(folder_volumes[0])
```

```python
#day 8
def read_data():
    path = 'data_day8.csv'
    with open(path) as f:
        data = f.read()
        return data.strip('\n\n')

data = read_data()
data = data.split('\n')
for index, line in enumerate(data):
    data[index] = list(line)
for index, line in enumerate(data):
    for index2, item in enumerate(line):
        line[index2] = int(item)


nb_col = len(data)
nb_ln = len(data[0])
nb_visible = 0
max_scenic_score = 0
for line in range(nb_col):
    for tree in range(nb_ln):
        list_north = [data[i][tree] for i in range(line)][::-1]
        list_south = [data[i][tree] for i in range(line+1, nb_col)]
        list_west = [data[line][i] for i in range(tree)][::-1]
        list_east = [data[line][i] for i in range(tree+1,nb_ln)]

        tree_height = data[line][tree]
        west_visible=len(list_west) == 0 or tree_height > max(list_west)
        east_visible=len(list_east) == 0 or tree_height > max(list_east)
        north_visible=len(list_north) == 0 or tree_height > max(list_north)
        south_visible=len(list_south) == 0 or tree_height > max(list_south)

        visible = west_visible or east_visible or north_visible or south_visible
        if visible == True:
            nb_visible += 1

        scenic_score = 1
        tree_lines = [list_north, list_south, list_west, list_east]
        for direction in tree_lines:
            if len(direction) == 0:
                scenic_score = scenic_score * 0
            elif True not in [height >= tree_height for height in direction]:
                scenic_score = scenic_score * len(direction)
            else:
                scenic_score = scenic_score * (list(height >= tree_height for height in direction).index(True)+1)
        max_scenic_score = max(max_scenic_score, scenic_score)
print(nb_visible)
print(max_scenic_score)
```

```python
#day 8
def read_data():
    path = 'data_day9.csv'
    with open(path) as f:
        data = f.read()
    data = data.strip('\n\n').split('\n')
    return data


data = read_data()

head_x_axis = 0
head_y_axis = 0
tail_x_axis = 0
tail_y_axis = 0
knots_position = [[0, 0] for i in range(1, 10)]
tail_positions = {(0, 0)}

for instruction in data:
    direction, amount = instruction.split(' ')
    amount = int(amount)
    for _ in range(amount, 0, -1):
        if direction == 'U':
            head_y_axis += 1
        elif direction == 'D':
            head_y_axis -= 1
        elif direction == 'L':
            head_x_axis -= 1
        elif direction == 'R':
            head_x_axis += 1
        
        for knot_number, position in enumerate(knots_position):
            tail_x_axis, tail_y_axis = position
            if knot_number == 0:
                prev_knot_x_axis = head_x_axis
                prev_knot_y_axis = head_y_axis
            else:
                prev_knot_x_axis = knots_position[int(knot_number)-1][0]
                prev_knot_y_axis = knots_position[int(knot_number)-1][1]

            y_diff = abs(prev_knot_y_axis - tail_y_axis)
            x_diff = abs(prev_knot_x_axis - tail_x_axis)

            if y_diff == 2:
                tail_y_axis = int((tail_y_axis + prev_knot_y_axis) / 2)
            elif y_diff == 1:
                if x_diff == 2:
                    tail_y_axis = prev_knot_y_axis
            if x_diff == 2:
                tail_x_axis = int((tail_x_axis + prev_knot_x_axis) / 2)
            elif x_diff == 1:
                if y_diff == 2:
                    tail_x_axis = prev_knot_x_axis
            knots_position[knot_number] = [tail_x_axis, tail_y_axis]
        
        tail_positions.add((knots_position[-1][0], knots_position[-1][1]))

print(len(tail_positions))
    
```
